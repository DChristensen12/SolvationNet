# Propylene Carbonate

```python
# Propylene Carbonate
density = 1.205 # in units of g/cm^3
molar_mass = 102.09 # in units of g/mol
NA = 6.022e23

# Setting the Box size
L_angstrom = 30
volume_A3 = L_angstrom**3 # Å^3
volume_cm3 = volume_A3 * 1e-24 #  cm^3

mass_g = density * volume_cm3 # The mass inside the box
moles = mass_g / molar_mass # The number of moles
num_molecules = moles * NA # Number of molecules

if print_it:
    print("Box volume (cm^3):", volume_cm3)
    print("Mass needed (g):", mass_g)
    print("Number of PC molecules:", num_molecules)
    print("Rounded:", round(num_molecules))
```

```python
# Creating the Propylene Carbonate Structure
%%bash
cat > pc.pdb << 'EOF'
HETATM    1  C1  PC      1       0.000   0.000   0.000
HETATM    2  O1  PC      1       1.200   0.000   0.000
HETATM    3  O2  PC      1      -1.200   0.000   0.000
HETATM    4  C2  PC      1       0.000   1.200   0.000
END
EOF
```

```python
%%bash
cat > input.inp << 'EOF'
tolerance 2.0
filetype pdb
output pc_box_30A.pdb

structure pc.pdb
  number 19
  inside box 0. 0. 0. 30. 30. 30.
end structure
EOF
```

# Adding Dimethoxyethane with the Propylene Carbonate

```python
density = 1.205  # g/cm^3 (assuming mixture density)
NA = 6.022e23

# Molecular weights
MW_PC = 102.09
MW_DME = 90.12

# This is the Box size
L = 30  # Angstrom
volume_A3 = L**3
volume_cm3 = volume_A3 * 1e-24

# Average molar mass (50:50 mole mixture)
MW_avg = (MW_PC + MW_DME) / 2

# total molecules
mass = density * volume_cm3
moles = mass / MW_avg
N_total = moles * NA

N_total = round(N_total)
N_PC = N_total // 2
N_DME = N_total - N_PC

if print_it:
  print("Total molecules:", N_total)
  print("PC molecules:", N_PC)
  print("DME molecules:", N_DME)
```
